```python
# 1. SETUP: Using a List to hold the entire inventory
inventory = []

# 2. DATA: Creating data using Dictionaries and Tuples

# Product 1
p1_info = ("P-101", "Laptop Pro", 1200.00, 5)  # Tuple (ID, Name, Price, Stock)
p1 = {
    "product_id": p1_info[0],
    "name": p1_info[1],
    "price": p1_info[2],
    "stock": p1_info[3]
}
inventory.append(p1)

# Product 2
p2_info = ("P-205", "Wireless Mouse", 25.50, 50) # Tuple (ID, Name, Price, Stock)
p2 = {
    "product_id": p2_info[0],
    "name": p2_info[1],
    "price": p2_info[2],
    "stock": p2_info[3]
}
inventory.append(p2)

# 3. OPERATION: Function to display the inventory and calculate total value
def display_inventory(inv_list):
    """Prints all products and calculates the total value of stock."""
    total_inventory_value = 0
    
    print("\n--- Current Inventory ---")
    print(f"Total Products in Stock: {len(inv_list)}\n")
    
    # Iterate over the List (inventory)
    for product in inv_list:
        # Access elements of the Dictionary (product)
        product_value = product["price"] * product["stock"]
        total_inventory_value += product_value
        
        # Display the product details
        print(f"ID: {product['product_id']} | Name: {product['name']}")
        print(f"  > Price: ${product['price']:.2f} | Stock: {product['stock']} units")
        print(f"  > Item Value: ${product_value:.2f}")

    print("\n-------------------------")
    print(f"TOTAL INVENTORY VALUE: ${total_inventory_value:.2f}")
    
# 4. EXECUTION
display_inventory(inventory)

# 5. DEMONSTRATION of List and Dictionary Mutability
print("\n--- Demonstration of Mutability ---")

# Accessing the second product in the List
mouse_product = inventory[1] 

# Updating the Dictionary value (Modifying Stock)
print(f"Original Mouse Stock: {mouse_product['stock']}")
mouse_product['stock'] = 45 # Dictionary is mutable
print(f"New Mouse Stock: {mouse_product['stock']}")

# 6. FINAL OUTPUT
display_inventory(inventory)
```
